```python
import matplotlib.pyplot as plt
import numpy as np

# fixed costs per year
fixed_costs = 15000

# variable costs per user per year
variable_costs_per_user = 10

# number of users on each plan
num_basic_users = 10
num_standard_users = 5
num_premium_users = 10

# target total annual profit
target_revenue = fixed_costs * 1.0

# initialize lists of basic, standard, and premium plan prices, and corresponding profits
basic_prices = []
standard_prices = []
premium_prices = []
total_revenues = []

# try different standard plan prices, starting at a suitable value
for i in range(20, 80):
    # calculate standard plan price and revenue
    standard_price = i*2
    standard_revenue = num_standard_users * standard_price
    
    # calculate basic plan price and revenue
    basic_price = 0.5 * standard_price
    basic_revenue = num_basic_users * basic_price
    
    # calculate premium plan price and revenue
    premium_price = 1.1 * standard_price
    premium_revenue = num_premium_users * premium_price
    
    # calculate total costs
    total_costs =  (num_basic_users + num_standard_users + num_premium_users) * 12 * variable_costs_per_user
    
    # total profit is the sum of revenues minus the total costs
    total_revenue = ((basic_revenue + standard_revenue + premium_revenue) * 12) - total_costs
    
    # add basic, standard, and premium prices, and total profit to lists
    basic_prices.append(basic_price)
    standard_prices.append(standard_price)
    premium_prices.append(premium_price)
    total_revenues.append(total_revenue)

# find the index of the point where total revenue is closest to the target revenue
differences = [abs(revenue - target_revenue) for revenue in total_revenues]
min_difference = min(differences)
index = differences.index(min_difference)

# plot the results
plt.plot(basic_prices, total_revenues)
plt.plot(standard_prices, total_revenues)
plt.plot(premium_prices, total_revenues)

# add horizontal line at y=30000
plt.hlines(y=target_revenue, xmin=0, xmax=250, color='red', linewidth=1, linestyle='--')
plt.text(x=110, y=target_revenue*1.06, s=f'Basic: {basic_prices[index]:.2f} \nStandard: {standard_prices[index]:.2f}  \nPremium: {premium_prices[index]:.2f}')

plt.xlabel('Plan Prices')
plt.ylabel('Total Annual Revenue')
plt.legend(['Basic', 'Standard', 'Premium', 'Target Revenue'], loc='upper right')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the base fee for each pricing plan
base_fees = [36, 72, 79]

# Define the additional charges per document for each pricing plan
additional_charges = [0.02, 0.01, 0.005]

# Define the number of documents per month for each pricing plan
documents_per_month_included = [7000, 12000, 20000]

# Calculate the unit price for each pricing plan at each number of documents per month
for i, base_fee in enumerate(base_fees):
    documents_per_month = range(1, 25000)

    unit_prices = []
    
    for n, doc in enumerate(documents_per_month):
        if doc > documents_per_month_included[i]:
            unit_prices.append(base_fee + additional_charges[i] * (n - documents_per_month_included[i]))
        else:
            unit_prices.append(base_fee)

    plt.plot(documents_per_month, unit_prices, label=f"Plan {i+1}")

# Add a legend and labels to the plot
plt.legend(['Basic, 7k included, 2 extra', 'Standard, 12k included, 1 extra', 'Premium, 20k included, .5 extra', 'Target Revenue'], loc='upper left')
plt.xlabel("Number of documents [/month]")
plt.ylabel("Price [/month]")

# Add a point to the plot
plt.plot(7000, 36, 'ro')

# Add a text label to the plot
plt.text(8900, 30, "Average number of documents\n for customer Boehnlein")

# draw a dotted, diagonal line connecting the point and the text
plt.plot([7000, 8700], [36, 36], 'r--')




# Show the plot
plt.show()
```
